#Exercise 5: Export A Dataframe To Excel Format And JSON Format

```python
simple_dataframe = pd.DataFrame({
       'Student': ['Yaron', 'Gila', 'Nurit', 'Aviva', 'Sharon'],
       'Age': ['47', '52', '46', '55', '47'],
       'City': ['Rehovot', 'Nes Ziona', 'Rehovot', 'Rehovot', 'Ashdod']
   })
```

```python
simple_dataframe.to_excel('Developers Institute.xlsx', sheet_name = 'Students')
```

```python
simple_dataframe.to_json('Developers Institute_split.json', orient = 'split')
```

```python
simple_dataframe.to_json('Developers Institute_records.json', orient = 'records')
```

```python
simple_dataframe.to_json('Developers Institute_index.json', orient = 'index')
```

```python
simple_dataframe.to_json('Developers Institute_columns.json', orient = 'columns')
```

```python
simple_dataframe.to_json('Developers Institute_values.json', orient = 'values')
```

Good to see all the different JSON structures!
