# Exercise 3: Identifying Data Types

```python
sleep_data.info()
```

```python
sleep_data.groupby('Age Group').mean()
```

Quantitative are the int64 and float64 datatypes, so Year, Avg hrs per day sleeping and Standard Error. This corresponds also with what is shown if you check the first rows with head().\
\
Qualitative would the object datatypes, so Period, Type of Days, Age Group, Activity, Sex. However Age Group is also partly quantitative, because the age groups are divided in bins of equal size.

```python
mental_health_data.info()
```

Quantitative are Year, Schizophrenia (%), Bipolar disorder (%), Eating disorders (%), Anxiety disorders (%), Drug use disorders (%), Depression (%), Alcohol use disorders (%), if you look at the data itself, not at the Dtype.

Qualitative are the object datatypes Entity, Code.

```python
clean_dataset.info()
```

```python
qualitative_columns = clean_dataset.select_dtypes(include=['object']).columns
print(qualitative_columns)
```

```python
quantitative_columns = clean_dataset.select_dtypes(include=['float64', 'int64']).columns
print(quantitative_columns)
```

Quantitative are Age, Debt, YearsEmployed, CreditScore, Income if you look at the data itself in the head(), not at the Dtype.

Qualitative are the object datatypes Industry, Ethnicity, Citizen, but also Gender, Married, Bankcustomer, Priordefault, Employed, Driverslicense, Zipcode, Approved. Many of those are boolean values. And Zipcodes, although they have a numerical appearance, don't have a numerically meaningful value.

```python
crx.info()
```

```python
qualitative_columns = crx.select_dtypes(include=['object']).columns
print(qualitative_columns)
```

```python
quantitative_columns = crx.select_dtypes(include=['float64', 'int64']).columns
print(quantitative_columns)
```

This is not a meaningful table. The headers are not meaningful.\
\
Looking at the five first rows with head(), the columns with headers b, ug, w, v, t, t.1, f, g.1, 00202, and + appear to be qualitative.\
\
And the columns with the headers 30.83, 0, 1.25, 01, 0.1 appear to be quantitative.

So, the conclusion of Exercise 3 is that the Dtype that pandas gives, is not as informative as one could wish. Lesson learned!
